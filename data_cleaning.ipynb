2.mengisi missing value                                            

menggunakan  fillna()untuk mengganti nilai kosong dengan nilai tertentu

2.1 Dengan nilai tertentu (seperti 0)


Kapan digunakan:

            -Data numerik
            Kolom mewakili jumlah/hitungan
            -Nilai kosong bermakna "tidak ada aktifitas"
            Contoh:Jumlah transaksi pelanggan, di mana NaN berarti tidak ada transaksi
            -Dataset tidak berisi koreksi kompleks
            Tidak ada hubungan mendalam yang bisa di ganggu dengan mengisi nilai kosong

Situasi:

        -Dataset e-commerce
         Kolom jumlah pembelian yang kosong
        -Data Survey
         kolom frekuensi aktititas dimana kosong berarti"tidak pernah di lakukan."

```python
df.fillna(0)# mengubah nilai NaN menjadi 0
```

```python
df['name'] = df['name'].fillna(0)#mengubah satu kolom saja menjadi 0
df
```

2.2 Foewwaed fill (ffill) dan Backward(bfill)

Kapan digunakan:

            -Data Time series / berurutan
              Data memiliki urutan logis seperti data waktu / lokasi geografis
              Nilai sebelumnya / selanjutnya relavan untuk mrngisi celah
            -Data tidak fluktuaktif
              Nilai antar waktu cenderung konsisten

Situasi:

      -Time series data
        Dataset penjualan harian
        Data cuaca harian di mana sensor gagal pada beberapa waktu tertentu
      -Data lokasi
        Dataset lokasi kendaraan yang hilang pada titik tertentu
        

```python
df['age'] = df['age'].ffill()#mendapatkan nilai yang kosong dari atas
df
```

```python
df['age'] = df['age'].bfill()#mendapatkan nilai yang kosong dari bawah
df
```

2.3 Mengisi missing value dengan mean,Median,dan Modus

1.Mengisi dengan Mean(rata-rata)

Kapan digunakan:

            -Data bersifat numerik
            -Data terdistribusi secara normal
            -Rata-rata merepresentasikan pusat data dengan baik

Contoh situasi:

      -Data nilai ujian siswa,dimana beberapa nilai hilang
      -Data penjualan produk dengan sedikit nilai yang hilang

2.Mengisi dengan median

Kapan digunakan:

            -Data bersifat secara numeric tetapi tidak terdistribusi secara normal(skewed)
            -Terdapat banyak outlier yang mempengaruhi rata-rata

Contoh situasi:

          -Data pendapatan masyarakat,di mana sebagian besar pendapatan berada di kategori menengah

          -Data durasi kunjungan ke situs web

3.Mengisi dengan modus

Kapan digunakan:

            -Data bersifat kategorikal/nominal
            -Kategori dengan frekuensi tertinggi di anggap paling mewakili data yang hilang

Contoh situasi:

        -Data preferensi pelanggan(misalnya, warna faforit/produk favorit)
        -Data demografis seperti gender / wilayah geografis
